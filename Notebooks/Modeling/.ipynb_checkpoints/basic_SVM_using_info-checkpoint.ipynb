## SVM on location

####  Get relevant information

```python
def get_coords(info):
    info = json.loads(info)
    lat = info['origins'][0]['latitude']
    lon = info['origins'][0]['longitude']
    mag = info['magnitudes'][0]['mag']
    event_type = info['event_type']
    return lat, lon, mag, event_type
```

```python
def get_all_coords_mag(info_set):
    data = np.empty((len(info_set), 3))
    labels = []
    for idx, info in enumerate(info_set):
        data[idx] = get_coords(info)[:3]
        labels.append(get_coords(info)[3])
    labels = np.array(labels)
    return data, labels
```

```python
def convert_labels(labels):
    labels_binary = np.empty((len(labels)), dtype=int)
    for idx, label in enumerate(labels):
        if label == "explosion":
            labels_binary[idx] = 0
        else:
            labels_binary[idx] = 1
    return labels_binary
```

```python
train_data, train_labels = get_all_coords_mag(train_info)
test_data, test_labels = get_all_coords_mag(test_info)
```

### Polynomial SVM

```python
def svm_model_selection(degree_range, X_train, y_train, X_test, y_test):
    best_score = 0.0
    best_model = None
    all_scores = []
    degrees = range(1, degree_range)
    for i in range(1, degree_range):
        model = SVC(kernel='poly', degree=i)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        score = model.score(X_test, y_test)
        all_scores.append(score)
        if score > best_score:
            best_score = score
            best_model = model
    
    plt.plot(degrees, all_scores)
    plt.xlabel('Degrees')
    plt.ylabel('Accuracy')
    plt.title('Mean accuracy of predict')
    plt.show()
    return best_model, best_score
        
```

## RBF SVM

source: https://scikit-learn.org/stable/auto_examples/svm/plot_rbf_parameters.html

```python
def rbf_SVM_model_selection_fine_tuning(c_range, gamma_range, X_train, y_train, X_test, y_test):
    best_model = None
    best_score = 0.0
    best_c = None
    best_gamma = None
    classifiers = []
    best_classifier = None
    all_scores = []
    for C in c_range:
        for gamma in gamma_range:
            model = SVC(kernel = 'rbf', C=C, gamma = gamma)
            model.fit(X_train, y_train)
            classifiers.append((C, gamma, model))
            y_pred = model.predict(X_test)
            score = model.score(X_test, y_test)
            all_scores.append(score)
            if score > best_score:
                best_score = score
                best_model = model
                best_c = C
                best_gamma = gamma
                best_classifier = (C, gamma, model)
    
    print(f'Best accuracy: {best_score}. Best C: {best_c}. Best gamma: {best_gamma}')
    return best_model, classifiers, all_scores, best_classifier
    
    
```

```python
C_range = np.logspace(-2, 10, 13)
gamma_range = np.logspace(-9, 3, 13)
param_grid = dict(gamma=gamma_range, C=C_range)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=cv)
grid.fit(X_train, y_train)
```

### Conclusions

The poly model correctly classifies both explosions and earthquake events fairly well. However, the model will not be able to correctly classify earthquake events that occur near mines nor explosion events that occur outside of mines. Good starting off point. Also as a continious line does not make the best classifier as there are regions with a high density of one class surrounded by a other classes.

The RBF model is a better choice for such data as it accounts for the clusters of one class within a cluster of a different class. The accuracy is higher, although the amount of data is too limited to make any definite conclusions. This classifier also suffers from misclassifying earthquakes near mines, and likely would misclassify explosions as earthquakes if an explosion occurs outside of the mines.
